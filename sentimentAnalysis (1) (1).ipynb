```python
from transformers import AutoTokenizer
from transformers import AutoModelForSequenceClassification
from scipy.special import softmax
import torch
```

```python
#getting RoBERTa
MODEL = f"cardiffnlp/twitter-roberta-base-sentiment"
tokenizer = AutoTokenizer.from_pretrained(MODEL)
model = AutoModelForSequenceClassification.from_pretrained(MODEL)
```

```python
#Returns polarity scores for RoBERTa
def polarity_scores_roberta(example):
    encoded_text = tokenizer(example, return_tensors='pt')
    output = model(**encoded_text)
    scores = output[0][0].detach().numpy()
    scores = softmax(scores)
    scores_dict = {
        'roberta_neg' : scores[0],
        'roberta_neu' : scores[1],
        'roberta_pos' : scores[2]
    }
    return scores_dict
```

```python
#running sentiment analysis for the control prompt AAVE answers
#gemini
control_aave['roberta_neg'] = None
control_aave['roberta_neu'] = None
control_aave['roberta_pos'] = None
for i, row in tqdm(control_aave.iterrows(), total=len(control_aave)):
    try:
        text = row['feedback']
        scores = polarity_scores_roberta(text)
        control_aave.at[i, 'roberta_neg'] = scores['roberta_neg']
        control_aave.at[i, 'roberta_neu'] = scores['roberta_neu']
        control_aave.at[i, 'roberta_pos'] = scores['roberta_pos']
    except RuntimeError:
        print(f'Broke for id {i}')
'''
#gpt
control_aave['roberta_neg'] = None
control_aave['roberta_neu'] = None
control_aave['roberta_pos'] = None
for i, row in tqdm(control_aave.iterrows(), total=len(control_aave)):
    try:
        text = row['gpt_feedback']
        scores = polarity_scores_roberta(text)
        control_aave.at[i, 'roberta_neg'] = scores['roberta_neg']
        control_aave.at[i, 'roberta_neu'] = scores['roberta_neu']
        control_aave.at[i, 'roberta_pos'] = scores['roberta_pos']
    except RuntimeError:
        print(f'Broke for id {i}')
'''
```
